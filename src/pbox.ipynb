```python
# """
# Apply bit permutation using a fixed mapping.
# :param value: 16-bit input value
# :return: Permuted 16-bit output
# """
def apply_pbox(value):
    permuted_value = 0
    for i in range(16):
        new_position = (i * 3) % 16
        bit = (value >> i) & 1
        permuted_value |= (bit << new_position)
    return permuted_value
```
