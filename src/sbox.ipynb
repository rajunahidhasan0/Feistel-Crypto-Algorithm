```python
# sbox.py
import random
# """ Generate a 4x4 S-Box with random permutation of 0-15 values. """
def generate_sbox():
    values = list(range(16))
    random.shuffle(values)
    return values


# """ Apply S-Box transformation on a 4-bit value. """
def apply_sbox(value, sbox):
    return sbox[value]

# """ Compute the inverse S-Box. """
def inverse_sbox(sbox):
    inv_sbox = [0] * 16
    for i in range(16):
        inv_sbox[sbox[i]] = i
    return inv_sbox
```
